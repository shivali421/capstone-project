```python
import sqlite3
```

```python
db=sqlite3.connect("women crime record.db")
```

```python
cursor=db.cursor()
```

```python
cursor.execute("""
    CREATE TABLE crimerecord (
        ID INT PRIMARY KEY,
        STATE_UT TEXT,
        DISTRICT TEXT,
        Year INT,
        Rape INT,
        Kidnapping_and_Abduction INT,
        Dowry_Deaths INT,
        Assault_on_women_with_intent_to_outrage_her_modesty INT,
        Insult_to_modesty_of_Women INT,
        Cruelty_by_Husband_or_his_Relatives INT,
        Importation_of_Girls INT
    )
""")
```

```python
with open('42_District_wise_crimes_committed_against_women_2001_2012.csv', 'r') as file:
    total_records = 0
    for row in file:
        # Assuming each row contains 11 values
        values = tuple(row.strip().split(","))
        cursor.execute("INSERT INTO crimerecord VALUES (?,?,?,?,?,?,?,?,?,?,?)", row.split(","))
        db.commit()
        total_records += 1
        print(total_records, 'Records inserted')
```

REQUIREMENT 3.2

Write SQL query to find the highest number of rapes & Kidnappings that happened in which state, District, and year

```python
# Find the highest number of rapes and kidnappings
result = cursor.execute("""
    SELECT STATE_UT, DISTRICT, Year, Rape, Kidnapping_and_Abduction
    FROM crimerecord
    WHERE Rape = (SELECT MAX(Rape) FROM crimerecord)
    OR Kidnapping_and_Abduction = (SELECT MAX(Kidnapping_and_Abduction) FROM crimerecord)
""")

# Print the result
for row in result:
    print(row)
```

REQUIREMENT 3.3

Write SQL query to find All the lowest number of rapes & Kidnappings that happened in which state, District, and year


```python
# Find the lowest number of rapes and kidnappings
result = cursor.execute("""
    SELECT STATE_UT, DISTRICT, Year, Rape, Kidnapping_and_Abduction
    FROM crimerecord
    WHERE Rape = (SELECT MIN(Rape) FROM crimerecord)
    OR Kidnapping_and_Abduction = (SELECT MIN(Kidnapping_and_Abduction) FROM crimerecord)
""")

# Print the result
for row in result:
    print(row)
```

REQUIREMENT 3.8.2

Write SQL query to find which District in each state/ut has the highest number of murders yearwise. Your output should show STATE/UT, YEAR, DISTRICT, and MURDERS.



```python
result = cursor.execute("""
    WITH RankedMurders AS (
        SELECT
            STATE_UT,
            YEAR,
            DISTRICT,
            MURDER,
            ROW_NUMBER() OVER (PARTITION BY STATE_UT, YEAR ORDER BY MURDER DESC) AS RowNum
        FROM murderrape
    )
    
    SELECT STATE_UT, YEAR, DISTRICT, MURDER
    FROM RankedMurders
    WHERE RowNum = 1
""")
for row in result:
    print(row)
```

```python
# Create a DataFrame
df = pd.DataFrame(query_result, columns=columns)
```

```python
# Display the DataFrame
print(df)
```

```python
# Count the number of appearances for each district
district_counts = df.groupby('DISTRICT')['Year'].count()

# Filter districts that appear 3 or more times
districts_3_or_more = district_counts[district_counts >= 3].index

# Filter the original DataFrame for these districts
filtered_df = df[df['DISTRICT'].isin(districts_3_or_more)]

# Sort the DataFrame by year in descending order
sorted_df = filtered_df.sort_values(by='Year', ascending=False)

# Display the result
print(sorted_df[['STATE/UT','YEAR','DISTRICT','MURDERS']])
```

```python
# Plotting a bar chart
plt.figure(figsize=(10, 6))
plt.bar(sorted_df['DISTRICT'], sorted_df['MURDER'], color='skyblue')
plt.xlabel('District')
plt.ylabel('Murders')
plt.title('Murders in Districts Appearing 3 or More Years')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plotting a line chart
plt.figure(figsize=(10, 6))
for district, query_result in sorted_df.groupby('DISTRICT'):
    plt.plot(query_result['Year'], query_result['MURDER'], label=district, marker='o')

plt.xlabel('Year')
plt.ylabel('Murders')
plt.title('Murders Over Years in Districts Appearing 3 or More Years')
plt.legend()
plt.show()
```
