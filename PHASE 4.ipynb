```python
import pandas as pd
```

```python
df=pd.read_csv('crime cluster.csv', sep='\t')
```

```python
df
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming data is stored in a DataFrame named 'df'
# Extracting relevant columns for clustering
#columns_for_clustering = ['MURDER', 'ATTEMPT TO MURDER', 'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER',
                           #'RAPE', 'KIDNAPPING & ABDUCTION', 'ROBBERY', 'RIOTS', 'CHEATING', 'THEFT', 'CRIMINAL BREACH OF TRUST']
```

```python
columns_for_clustering = ['TOTAL CRIMES']
```

```python
# Creating a new DataFrame for clustering
cluster_data = df[columns_for_clustering]
```

```python
# Standardizing the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data)
```

```python
# Applying K-means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['Crime_Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
# Displaying the resulting DataFrame with cluster assignments
print(df[['STATE/UT', 'DISTRICT', 'TOTAL CRIMES', 'Crime_Cluster']])
```

REQUIREMENT 4.2

```python
# Separate DataFrames for each cluster
cluster_0_df = df[df['Crime_Cluster'] == 0]
cluster_1_df = df[df['Crime_Cluster'] == 1]
cluster_2_df = df[df['Crime_Cluster'] == 2]
```

```python
# Displaying the separate DataFrames
print("Sensitive Areas (Cluster 0):")
print(cluster_0_df[['STATE/UT', 'DISTRICT', 'TOTAL CRIMES']])
```

```python
print("\nModerate Areas (Cluster 1):")
print(cluster_1_df[['STATE/UT', 'DISTRICT', 'TOTAL CRIMES']])
```

```python
print("\nPeaceful Areas (Cluster 2):")
print(cluster_2_df[['STATE/UT', 'DISTRICT', 'TOTAL CRIMES']])
```
