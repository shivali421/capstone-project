```python
# Encode the categorical column 'States & UTs'
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2['States & Uts Encoded'] = le.fit_transform(df2['States & Uts'])
```

```python
# Sort the DataFrame by Average Literacy Rate for better visualization
df_sorted = df2.sort_values(by='Average Literacy Rate %', ascending=False)
```

```python
# Bar chart for Average Literacy Rate and Total Crimes
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 12))

# Bar chart for Average Literacy Rate
sns.barplot(x='Average Literacy Rate %', y='States & Uts Encoded', data=df_sorted, palette='viridis', ax=axes[0])
axes[0].set_yticks(df_sorted['States & Uts Encoded'])
axes[0].set_yticklabels(le.inverse_transform(df_sorted['States & Uts Encoded']))
axes[0].set_xlabel('Average Literacy Rate (%)')
axes[0].set_ylabel('States & Uts')
axes[0].set_title('Average Literacy Rate across States & Uts')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the "States & UTs" column
df2['States & Uts Encoded'] = label_encoder.fit_transform(df2['States & Uts'])

# Display the encoded values
encoded_states = df2[['States & Uts', 'States & Uts Encoded']].drop_duplicates()
print(encoded_states)
```

```python
# Extract relevant columns for correlation analysis
crime_columns = ['MURDER',
       'ATTEMPT TO MURDER', 'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER',
       'RAPE', 'CUSTODIAL RAPE', 'OTHER RAPE', 'KIDNAPPING & ABDUCTION',
       'KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS',
       'KIDNAPPING AND ABDUCTION OF OTHERS', 'DACOITY',
       'PREPARATION AND ASSEMBLY FOR DACOITY', 'ROBBERY', 'BURGLARY', 'THEFT',
       'AUTO THEFT', 'OTHER THEFT', 'RIOTS', 'CRIMINAL BREACH OF TRUST',
       'CHEATING', 'COUNTERFIETING', 'ARSON', 'HURT/GREVIOUS HURT',
       'DOWRY DEATHS', 'ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY',
       'INSULT TO MODESTY OF WOMEN', 'CRUELTY BY HUSBAND OR HIS RELATIVES',
       'IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES',
       'CAUSING DEATH BY NEGLIGENCE', 'OTHER IPC CRIMES']

crime_df = df2[['Average Literacy Rate %'] + crime_columns]

# Convert relevant columns to numeric
crime_df_numeric = crime_df.apply(pd.to_numeric, errors='coerce')
# Calculate the correlation matrix
correlation_matrix = crime_df_numeric.corr()

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Crime vs Literacy Rate')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Create a box plot to identify outliers
sns.boxplot(x=df2['Average Literacy Rate %'], y=df2['TOTAL CRIMES'], palette='Set2')

plt.title('Box Plot of Literacy Rate vs Total Crimes')
plt.show()
```

```python
from scipy import stats

# Calculate Z-scores for 'Average Literacy Rate %' and 'TOTAL CRIMES'
z_scores = stats.zscore(df2[['Average Literacy Rate %', 'TOTAL CRIMES']])

# Define a threshold for considering a point as an outlier (e.g., Z-score > 3)
threshold = 3

# Identify outlier indices
outlier_indices = (np.abs(z_scores) > threshold).any(axis=1)

# Get the rows corresponding to outliers
outliers = df2[outlier_indices]

# Print or analyze the outliers
print("Outliers:")
print(outliers)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# Create a joint plot with regression lines
sns.jointplot(x='Area(sq km)', y='TOTAL CRIMES', data=df4, kind='reg', height=8, joint_kws={'line_kws':{'color':'red'}})

plt.title('Relationship Between Area and Total Crimes by State')
plt.show()
```
